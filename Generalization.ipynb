```python
import gin
import sys
import os

def check_gin_file(file_path):
    """Attempts to parse a gin file and reports any errors."""
    if not os.path.isfile(file_path):
        print(f"Error: File not found at '{file_path}'")
        sys.exit(1)

    try:
        # The act of parsing the file will raise an exception if there is an error
        gin.parse_config_file(file_path)
        print(f"Success: '{file_path}' parsed without errors.")
    except Exception as e:
        print(f"Error: Failed to parse '{file_path}'")
        print("Detailed traceback:")
        # Print the detailed exception provided by gin-config
        print(e)
        sys.exit(1)

if __name__ == "__main__":
    # Replace 'path/to/your/config.gin' with the actual path to your file
    # You can also pass the file path as a command-line argument
    config_file = 'configs/TakeCover.gin'
    check_gin_file(config_file)
```
